```python
import os

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

from utils import *
```

```python
plt.rc('font', family='NanumGothic')
```

```python
df = build_dataframe('./20221111')
```

```python
evt = get_event_date()
```

```python
print(df.head().to_markdown())
```

# 날자에 따른 글 젠량

```python
fig = plt.figure(figsize=(36, 9), dpi=100)
g = sns.countplot(data=df, x='time')
g.tick_params(axis='x', rotation=45)
g.legend(*g.get_legend_handles_labels(), ncol=4)
g.bar_label(g.containers[0], labels=[evt.get(txt.get_text(), "") for txt in g.get_xticklabels()])
plt.tight_layout()
plt.savefig('count_all.svg')
plt.show()
```

```python
fig = plt.figure(figsize=(36, 9), dpi=100)
g = sns.countplot(data=df, x='time', hue='category')
g.tick_params(axis='x', rotation=45)
g.legend(*g.get_legend_handles_labels(), ncol=4)
g.bar_label(g.containers[0], labels=[evt.get(txt.get_text(), "") for txt in g.get_xticklabels()])
plt.tight_layout()
plt.savefig("count_cat.svg")
plt.show()
```

```python
fig = plt.figure(figsize=(36, 9), dpi=100)
g = sns.countplot(data=df, x='time', hue='job')
g.tick_params(axis='x', rotation=45)
g.legend(*g.get_legend_handles_labels(), ncol=4, loc='upper right')
g.bar_label(g.containers[0], labels=[evt.get(txt.get_text(), "") for txt in g.get_xticklabels()])
plt.tight_layout()
plt.savefig("count_job.svg")
plt.show()
```
