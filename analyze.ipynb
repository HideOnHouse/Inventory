```python
import os

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
for path, dirs, files in os.walk("./20221111/"):
    for file in files:     
        with open(path + os.sep + file, encoding='utf-8') as f:
            header = f.readline().rstrip().split('\t')
            buffer = []
            data = f.read().split('\t')
            for i in range(0, len(data), 4):
                row = [i.strip() for i in data[i:i+4]]
                if len(row) == 1:
                    buffer[-1].insert(0, row)  
                    continue
                elif i != 0:
                    reco = row[0].split('[')[0]
                    row[0] = row[0][row[0].index('['):]
                    buffer[-1].insert(-1, reco)
                buffer.append(row)
            res = pd.DataFrame(buffer, columns=header)
        res.to_csv(path + os.sep + file, sep='\t', index=False)
```
